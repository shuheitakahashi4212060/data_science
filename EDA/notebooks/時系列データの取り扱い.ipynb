```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = './drive/MyDrive/pandas-dataset/sales_transactions.xlsx'
```

```python
df = pd.read_excel(file_path, index_col=0)
```

```python
df
```

```python
df.info()
```

```python
df['ymd'] = df['ymd'].astype(str)
```

```python
df.dtypes
```

```python
df['ymd'] = pd.to_datetime(df['ymd'])
```

```python
df = df.fillna({'quantity': 1})
```

```python
df
```

```python
df['sales'] = df['revenue'] * df['quantity']
```

```python
df
```

```python
daily_sales_df = df.groupby('ymd')['sales'].sum().reset_index()
```

```python
daily_sales_df
```

```python
plt.plot(daily_sales_df['ymd'],daily_sales_df['sales'])
plt.xticks(rotation=90)
```
